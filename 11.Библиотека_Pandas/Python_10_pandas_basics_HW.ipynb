### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
rating_movie = pd.read_csv('ml-20m/ratings.csv')
name_movie = pd.read_csv('ml-20m/movies.csv')
joined_table = rating_movie.merge(name_movie, on='movieId', how='left')
joined_table.head()
```

```python
name_movie_rate_5 = joined_table[joined_table.rating == 5.0].title.value_counts()
print('Больше всего оценок 5.0 выставлено фильму:', name_movie_rate_5.index[0], '(кол-во оценок: ', name_movie_rate_5[0], ')')
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.info()
data.head()
```

```python
filtered_countries = data[(data.country.isin(['Latvia','Lithuania','Estonia'])) & (data.category.isin([4, 12, 21])) & (data.year >= 2005) & (data.year <= 2010) & (data.quantity >= 0) ]
sum_quantity = filtered_countries.quantity.sum()
print('Cуммарное потребление стран Прибалтики, категорий 4, 12 и 21 за период с 2005 по 2010 : ',sum_quantity)
```
