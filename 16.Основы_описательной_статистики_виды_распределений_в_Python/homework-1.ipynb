## Задание 1

Изобразите с помощью matplotlib. Изобразите точку x + 2y + 3z. Найдите угол между векторами x, y и x, z. 

```python
x = np.array([1,1])
y = np.array([2, 0])
z = np.array([0,2])
```

```python
a = x + 2*y + 3*z
print(a)
plt.plot(a[0],a[1], marker='.', markersize=12)
```

```python
def cosine( x, y ):
    """
    Подсчет косинуса угла между векторами x, y по их координатам
    """
    xLength = np.sqrt( (x*x).sum() ) #подсчет длины вектора x
    yLength = np.sqrt( (y*y).sum() ) #подсчет длины вектора y
    
    return np.dot( x, y ) / ( xLength * yLength )
```

```python
angle_x_y = round(np.arccos( cosine( x, y ) ) * 180 / np.pi)
print(f'Угол между векторами x,y в градусах: {angle_x_y}')
angle_x_z = round(np.arccos( cosine( x, z ) ) * 180 / np.pi)
print(f'Угол между векторами x,z в градусах: {angle_x_z}')
```

## Задание 3

Найдите сингулярное разложение матрицы.

```python
m3 = np.array([[1, 2],
               [2, 3]])
```

```python
U, s, V = np.linalg.svd(m3)
s = np.diag(s) 
print(U)
print(s)
print(V)
```

```python
m3_check = U.dot(s).dot(V) 
np.allclose(m3, m3_check) #Проверка
```
