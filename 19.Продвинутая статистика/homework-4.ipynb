Как видно по матрице корреляций, атрибут MEDV (средняя стоимость дома) имеет прямую зависимость с атрибутом RM (среднее кол-во комнат в доме) и обратную зависимость с атрибутом LSTAT(процент населения с низким уровнем жизни).То есть, чем больше комнат в доме, тем дороже дом, и чем дороже дом, тем меньше людей с низким урвнем жизни в нем проживают.

Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
lstat = np.array(data['LSTAT']).reshape(-1,1) #независивые переменные колонка LSTAT
med_value = np.array(data['MEDV']).reshape(-1,1) #зависимая переменна колонка MEDV
# параметр fit_reg - строит линию регрессии для двух выборок
sns.regplot(x= lstat, y=med_value, data=data, fit_reg=True)
plt.xlabel('Процент населения с низким уровнем жизни')
plt.ylabel('Средняя цена дома')
plt.title('Корреляционная зависимость двух выборок')
plt.grid(False)
plt.show()
```

```python
#Создаем класс модель
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#Разделим на 80% тренировочных и на 20% тестовых
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    lstat,
    med_value,
    test_size = 0.2
)
```

```python
#Обучим модель 
model.fit(X_train, y_train)
```

```python
#Посмотрим предсказания модели на тестируемых данных
predicted_y = model.predict(X_test)
```

```python
#Посчитаем ошибку
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(mean_squared_error(predicted_y, y_test))
print(mean_absolute_error(predicted_y, y_test))
```

```python
plt.xlabel('Процент населения с низким уровнем жизни')
plt.ylabel('Средняя цена дома')
plt.title('Линейная регрессия от одной переменной')
# Построим прямую для регрессии
# Создадим 100 точек для предсказаний
x = np.linspace(0, 40, 60)
# Посчитаем предсказанные значения
# y = kx + b
# k = model.coef_
# b = model.intercept_
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(0, 40)
plt.ylim(0, 60)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

В данном случае получается зависимость двух переменных: **чем дороже дом, тем меньше в нем людей с низким уровнем жизни.**

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
import random
number_of_test = 10 #10 подбрасываний играньного кубика
list_of_outcomes = []
for i in range(number_of_test):
    list_of_outcomes.append(random.randint(1, 6))
plt.hist(list_of_outcomes)
plt.xlabel('X values')
plt.ylabel('Number of elements in each X')
plt.xlim(0, 7)
plt.show()
```

Можно предположить, что это равномерное распеределение, что логично, потому что при подбрасывании одного честного кубика вероятность выпадения одной из 6 граней состовляет 1/6, то есть любое из значений выпадает с равной вероятностью 2,78%.
При увеличении кол-ва подбрасываний видно, что гистограмма приближается к виду равномерного распередения.

```python
number_of_test = 10000 #10 000 подбрасываний играньного кубика
list_of_outcomes = []
for i in range(number_of_test):
    list_of_outcomes.append(random.randint(1, 6))
plt.hist(list_of_outcomes)
plt.xlabel('X values')
plt.ylabel('Number of elements in each X')
plt.show()
```

Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
number_of_sum = 1000
number_of_test = 100
list_of_outcomes = []
sum_ = 0
for i in range(number_of_sum):
    for a in range(number_of_test):
        sum_ += random.randint(1, 6)
    list_of_outcomes.append(sum_)
    sum_ = 0
plt.hist(list_of_outcomes)
plt.xlabel('X values')
plt.ylabel('Number of elements in each X')
plt.show()
```

Распеделение стало нормальным. Чтобы понять почему, рассмотрим пример с 2-мя кубиками, сумма на гранях 2-х кубиков может изменяться от 2 до 12, причем 2 может получится только при одном исходе: 1+1, 12 тоже модет быть тольо при одном исходе: 6+6, значит вероятность выпадения 2 или 12 равна 2,78 и 2,78 соотвественно. Чтобы получить 4 есть  уже больше вариантов: 1+3, 2+2, 1+3, значит вероятность уже больше и равно 2,78*3 = 8,33 %. Самое большое кол-во комбинаций существует, чтобы получить 7, а это как раз верхушка распеределения и среднее значение выборки из сумм бросков 2-х кубиков. Для 100 кубиков такая же логика.
Применить ЦПТ можно, потому что исходы при броске игрального кубика являются независимыми, то есть, вероятность наступление одного исхода не изменяет вероятность наступления другого.
